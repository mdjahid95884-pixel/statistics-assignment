Question 1:What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

When we do hypothesis testing, we decide whether to reject the null hypothesis or not.
During this process, two kinds of mistakes can happen.
1. Type I Error (False Positive)
This happens when we reject the null hypothesis even though it is actually true.
It is like saying something is happening when it actually isn’t.
Example:
A medical test says a healthy person is sick.
Probability of Type I error = (alpha).
2. Type II Error (False Negative)
This happens when we fail to reject the null hypothesis even though it is false.
This means we miss something important.
Example:
A test shows a sick person is healthy.
Probability of Type II error = (beta).
Impact on decision-making:
If avoiding false alarms is more important reduce Type I error.
If avoiding missed detections is more important reduce Type II error.
Different applications prioritize different errors, like medicine or law.





Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

The P-value tells us how likely it is to get the observed sample result assuming the null hypothesis is true.
Interpretation:
Small P-value (< 0.05):
The result is unlikely under Ho → Reject Ho.
Large P-value (> 0.05):
The result is normal under Ho → Do not reject Ho.
simple meaning:
P-value measures the strength of evidence against Ho.




Question 3:Explain the difference between a Z-test and a T-test, including when to use each.

Both tests compare sample mean with the population mean, but they are used in different situations.
Z-Test
Use when:
Population standard deviation is known
Sample size is large (n > 30)
T-Test
Use when:
Population standard deviation is unknown
Sample size is small (n ≤ 30)
The t-distribution is wider to account for extra uncertainty in small samples.
Simple rule:
Large sample → Z-Test
Small sample → T-Test




Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

A confidence interval (CI) gives a range of values where the true population mean is expected to lie with a certain confidence level (like 95%).
Example:
95% CI = (12, 18)
This means we are 95% confident the real mean is between 12 and 18.
Margin of error affects width:
Bigger margin of error → wider interval (less precise)
Smaller margin of error → narrow interval (more precise)
Margin of error increases with:
Smaller sample
More variation in data
Higher confidence level




Question 5:Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

ANOVA (Analysis of Variance) is used to compare means of three or more groups at the same time.
Assumptions:
Samples are independent
Data is normally distributed
All groups have equal variances
How ANOVA extends hypothesis testing:
A t-test compares only two groups.
ANOVA compares multiple groups in one test and tells if at least one group mean is different.
If ANOVA is significant, we can do further tests to find which group differs.

```python
# Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

import numpy as np
from scipy.stats import norm

data = [12, 14, 11, 15, 13, 16, 14]
population_mean = 10
population_sd = 3

sample_mean = np.mean(data)
n = len(data)

z_score = (sample_mean - population_mean) / (population_sd / np.sqrt(n))
p_value = 1 - norm.cdf(z_score)

print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation:
# If P-value < 0.05 → reject H₀ → sample mean is significantly different from population mean.
```

```python
# Question 8:enerate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

import numpy as np
import matplotlib.pyplot as plt

samples = []

for i in range(1000):
    dist = np.random.exponential(scale=2, size=50)
    samples.append(np.mean(dist))

plt.hist(samples, bins=30)
plt.title("CLT Demonstration Using Exponential Distribution")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.show()

# This shows that means of non-normal data become approximately normal.
```

```python
# Question 9:Write a Python function to calculate and visualize the confidence interval for a sample mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sd = np.std(data, ddof=1)
    n = len(data)

    z = norm.ppf((1 + confidence) / 2)
    margin = z * (sd / np.sqrt(n))

    lower = mean - margin
    upper = mean + margin

    print("Confidence Interval:", (lower, upper))

    plt.axvline(lower, linestyle='--')
    plt.axvline(upper, linestyle='--')
    plt.hist(data)
    plt.show()

data = np.random.randn(40)
confidence_interval(data)
```

```python
# Question 10:Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

observed = [50, 30, 20]
expected = [40, 40, 20]

chi_stat, p_value = chisquare(observed, expected)

print("Chi-square value:", chi_stat)
print("P-value:", p_value)

# Explanation:
# If P-value < 0.05, the observed data does not fit the expected distribution.
```
